## Garden Groups - Part 1

Why not search for the Chief Historian near the gardener and his massive farm? There's plenty of food, so The Historians grab something to eat while they search.

You're about to settle near a complex arrangement of garden plots when some Elves ask if you can lend a hand. They'd like to set up fences around each region of garden plots, but they can't figure out how much fence they need to order or how much it will cost. They hand you a map (your puzzle input) of the garden plots.

Each garden plot grows only a single type of plant and is indicated by a single letter on your map. When multiple garden plots are growing the same type of plant and are touching (horizontally or vertically), they form a **region**. For example:

```
AAAA
BBCD
BBCC
EEEC
```

This 4x4 arrangement includes garden plots growing five different types of plants (labeled A, B, C, D, and E), each grouped into their own region.

In order to accurately calculate the cost of the fence around a single region, you need to know that region's **area** and **perimeter**.

The **area** of a region is simply the number of garden plots the region contains. The above map's type A, B, and C plants are each in a region of area 4. The type E plants are in a region of area 3; the type D plants are in a region of area 1.

Each garden plot is a square and so has **four sides**. The **perimeter** of a region is the number of sides of garden plots in the region that do not touch another garden plot in the same region. The type A and C plants are each in a region with perimeter 10. The type B and E plants are each in a region with perimeter 8. The lone D plot forms its own region with perimeter 4.

Visually indicating the sides of plots in each region that contribute to the perimeter using - and |, the above map's regions' perimeters are measured as follows:

```
+-+-+-+-+
|A A A A|
+-+-+-+-+     +-+
              |D|
+-+-+   +-+   +-+
|B B|   |C|
+   +   + +-+
|B B|   |C C|
+-+-+   +-+ +
          |C|
+-+-+-+   +-+
|E E E|
+-+-+-+
```

Plants of the same type can appear in multiple separate regions, and regions can even appear within other regions. For example:

```
OOOOO
OXOXO
OOOOO
OXOXO
OOOOO
```

The above map contains **five** regions, one containing all of the O garden plots, and the other four each containing a single X plot.

The four X regions each have area 1 and perimeter 4. The region containing 21 type O plants is more complicated; in addition to its outer edge contributing a perimeter of 20, its boundary with each X region contributes an additional 4 to its perimeter, for a total perimeter of 36.

Due to "modern" business practices, the **price** of fence required for a region is found by **multiplying** that region's area by its perimeter. The **total price** of fencing all regions on a map is found by adding together the price of fence for every region on the map.

In the first example, region A has price 4 * 10 = 40, region B has price 4 * 8 = 32, region C has price 4 * 10 = 40, region D has price 1 * 4 = 4, and region E has price 3 * 8 = 24. So, the total price for the first example is **140**.

In the second example, the region with all of the O plants has price 21 * 36 = 756, and each of the four smaller X regions has price 1 * 4 = 4, for a total price of **772** (756 + 4 + 4 + 4 + 4).

### Solve the problem

````python
let N = lines.Length
N
````

````python
let gardenMap: char list list = lines |> List.rev |> List.map (fun line -> line |> Seq.toList)
gardenMap[0]
````

````python
type Region = { plantType: char; points: (int*int) Set }

let area (region: Region): int = region.points |> Set.count

let isEdgePoint (region: Region) (point: int*int): bool =
    match point with
    | (col, row) when col = 0 || col = N-1 || row = 0 || row = N-1 -> true
    | (col, row) when not (
            (region.points |> Set.contains (col-1, row)) &&
            (region.points |> Set.contains (col+1, row)) &&
            (region.points |> Set.contains (col, row-1)) &&
            (region.points |> Set.contains (col, row+1))
        ) -> true
    | _ -> false

let countExposedEdges (region: Region) (point: int*int): int =
    let (col, row) = point
    [
        col = 0 || not (region.points |> Set.contains (col-1, row)) ;
        col = N-1 || not (region.points |> Set.contains (col+1, row)) ;
        row = 0 || not (region.points |> Set.contains (col, row-1)) ;
        row = N-1 || not (region.points |> Set.contains (col, row+1))
    ] |> List.filter (fun boolVal -> boolVal) |> List.length

let perimeter (region: Region): int =
    region.points |> Set.toSeq |> Seq.map (fun point -> countExposedEdges region point) |> Seq.sum
````

So, having done all of the above, how can we isolate the various regions in the garden?  A possible solution is:
* Create a region for each point in the garden, and include in that region all neighbouring points that have the same plant type (excluding diagonal neighbours).
* Unify any regions that have the same plant type and overlap (i.e. share any of the same points).

````python
let neighbours (gardenMap: char list list) (plantType: char) (point: int*int): (int*int) Set =
    let (col, row) = point
    (seq {
        if (col > 0 && gardenMap[row][col-1] = plantType) then yield (col-1, row)
        if (col < (N-1) && gardenMap[row][col+1] = plantType) then yield (col+1, row)
        if (row > 0 && gardenMap[row-1][col] = plantType) then yield (col, row-1)
        if (row < (N-1) && gardenMap[row+1][col] = plantType) then yield (col, row+1)
    }) |> Set.ofSeq

let rec unifyRegions (regions: Region list): Region list =
    match regions with
    | [] -> []
    | head::[] -> regions
    | head::tail ->
        let (unionTail, remainder) = tail |> List.partition (fun reg -> reg.plantType = head.plantType && not (Set.intersect head.points reg.points).IsEmpty)
        match unionTail with
        | [] -> head::(unifyRegions tail)
        | _ ->
            let allPoints = Set(seq{ for pt in head.points do yield pt ; for reg in unionTail do for pt in reg.points do yield pt} )
            unifyRegions ({ plantType = head.plantType; points = allPoints }::remainder)

let buildRegions (gardenMap: char list list): Region list =
    let startRegions =
        (seq {
            for col in 0..(N-1) do
                for row in 0..(N-1) do
                    let plantType = gardenMap[row][col]
                    yield { plantType = plantType; points = (neighbours gardenMap plantType (col,row)) |> Set.add (col, row) }
        }) |> Seq.toList
    startRegions |> unifyRegions
    
````

````python
let gardenRegions = buildRegions gardenMap
gardenRegions
````

````python
let totalPrice = gardenRegions |> List.map (fun region -> (area region) * (perimeter region)) |> List.sum
totalPrice
````

### Solve the problem

So how can we calculate the number of sides of each region?  It won't be as simple as the trick we used to calculate the perimeter.  We can
* Isolate the points that are on the boundary (edge) of each region.
* Calculate the edges for each of those edge points.
  * While point indices run from 0..N-1 in each direction, edge indices run 0..N in each direction.
* Sort edges into order, then take the first edge as the start of a side, look for other edges that complete that side, etc.
* **NOTE** - after some experimentation, it was discovered that a 'side' is more than just a contiguous set of edges in the same direction, it *also* has to have the points (garden blocks) on the *same side* of the whole edge.  This changes the result that you would otherwise get.

````python
type EdgeDirection = Horizontal | Vertical

type Edge = { point: int*int; direction: EdgeDirection; edgeStart: int*int; edgeEnd: int*int }

let edges (region: Region) (point: int*int): Edge Set =
    if (isEdgePoint region point) then
        let (col, row) = point
        (seq {
            if (col = 0) then yield {point = point; direction = Vertical; edgeStart = (0, row); edgeEnd = (0, row+1)}
            if (col = (N-1)) then yield {point = point; direction = Vertical; edgeStart = (N, row); edgeEnd = (N, row+1)}
            if (row = 0) then yield {point = point; direction = Horizontal; edgeStart = (col, 0); edgeEnd = (col+1, 0)}
            if (row = (N-1)) then yield {point = point; direction = Horizontal; edgeStart = (col, N); edgeEnd = (col+1, N)}
            if (col > 0 && not (region.points |> Set.contains (col-1, row))) then yield {point = point; direction = Vertical; edgeStart = (col, row); edgeEnd = (col, row+1)}
            if (col < N-1 && not (region.points |> Set.contains (col+1, row))) then yield {point = point; direction = Vertical; edgeStart = (col+1, row); edgeEnd = (col+1, row+1)}
            if (row > 0 && not (region.points |> Set.contains (col, row-1))) then yield {point = point; direction = Horizontal; edgeStart = (col, row); edgeEnd = (col+1, row)}
            if (row < N-1 && not (region.points |> Set.contains (col, row+1))) then yield {point = point; direction = Horizontal; edgeStart = (col, row+1); edgeEnd = (col+1, row+1)}
        } |> Set.ofSeq)
    else Set.empty
````

````python
let sortEdgesIntoSides (edges: Edge list): Edge list list =
    let horizontalEdges = edges |> List.filter (fun edge -> edge.direction = Horizontal) |> List.sortBy (fun edge -> edge.edgeStart)
    let verticalEdges = edges |> List.filter (fun edge -> edge.direction = Vertical) |> List.sortBy  (fun edge -> edge.edgeStart)
    let rec collectEdgesIntoSides (sides: Edge list list) (edges: Edge list): Edge list list =
        match edges with
        | [] -> sides
        | head::tail ->
            match sides with
            | [] -> collectEdgesIntoSides [[head]] tail
            | side::rest ->
                let sideLastEdge = side |> List.last
                let nextEdgeList, otherEdgesList = edges |> List.partition (fun edge ->
                    edge.edgeStart = sideLastEdge.edgeEnd && (
                        (edge.direction = Vertical && (fst edge.point) = (fst sideLastEdge.point)) ||
                        (edge.direction = Horizontal && (snd edge.point) = (snd sideLastEdge.point))
                    )
                )
                match nextEdgeList with
                | [] ->
                    let sides = collectEdgesIntoSides rest edges
                    side::sides
                | _ ->
                    collectEdgesIntoSides ((side @ nextEdgeList)::rest) otherEdgesList
    (seq {
        yield! (collectEdgesIntoSides [] horizontalEdges)
        yield! (collectEdgesIntoSides [] verticalEdges)
    } |> Seq.toList)

let sides (region: Region) =
    let edgePoints = region.points |> Set.filter (fun point -> isEdgePoint region point)
    let regionEdgeList = edgePoints |> Set.map (fun point -> edges region point) |> Set.toList |> List.fold Set.union Set.empty |> Set.toList
    let sides = sortEdgesIntoSides regionEdgeList
    sides

let sideCount (region: Region) = sides region |> List.length
````

````python
gardenRegions |> List.map (fun region -> (region.plantType, area region, sideCount region))
````

````python
let reducedPrice = gardenRegions |> List.map (fun region -> (area region) * (sideCount region)) |> List.sum
reducedPrice
````
